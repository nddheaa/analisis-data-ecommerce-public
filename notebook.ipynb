#### 1. `customers_df`
- **Deskripsi**: Dataset ini mencatat informasi tentang pelanggan.

### Cleaning Data

#### Membersihkan dataset yang duplikat

```python
# Memeriksa dan menghapus duplikasi di setiap dataset yang memiliki duplikasi

# Untuk customers_df
if customers_df.duplicated().sum() > 0:
    customers_df.drop_duplicates(inplace=True)

# Untuk geolocation_df
if geolocation_df.duplicated().sum() > 0:
    geolocation_df.drop_duplicates(inplace=True)

# Untuk order_items_df
if order_items_df.duplicated().sum() > 0:
    order_items_df.drop_duplicates(inplace=True)

# Untuk order_payments_df
if order_payments_df.duplicated().sum() > 0:
    order_payments_df.drop_duplicates(inplace=True)

# Untuk order_reviews_df
if order_reviews_df.duplicated().sum() > 0:
    order_reviews_df.drop_duplicates(inplace=True)

# Untuk product_df
if product_df.duplicated().sum() > 0:
    product_df.drop_duplicates(inplace=True)

# Untuk sellers_df
if sellers_df.duplicated().sum() > 0:
    sellers_df.drop_duplicates(inplace=True)
```

#### Mengganti tipe data menjadi datetime

```python
dt_order_item = ['shipping_limit_date']
for column in dt_order_item:
    order_items_df[column] = pd.to_datetime(order_items_df[column])
```

```python
order_items_df.info()
```

```python
dt_order_review = ['review_creation_date', 'review_answer_timestamp']
for column in dt_order_review:
    order_reviews_df[column] = pd.to_datetime(order_reviews_df[column])
```

```python
order_reviews_df.info()
```

```python
dt_orders = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for column in dt_orders:
    orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```

### RFM

```python
# Menentukan tanggal terakhir dalam dataset untuk menghitung Recency
latest_date = all_df['order_purchase_timestamp'].max().date()

# Menghitung Recency, Frequency, dan Monetary
rfm_df = all_df.groupby(by="customer_id", as_index=False).agg({
    "order_purchase_timestamp": "max",  # mengambil tanggal order terakhir
    "order_id": "count",                # menghitung jumlah order
    "payment_value": "sum"              # menghitung jumlah revenue yang dihasilkan
})

# Mengganti nama kolom
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

# Menghitung Recency
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (latest_date - x).days)

# Menghapus kolom 'max_order_timestamp' yang tidak diperlukan lagi
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

# Menampilkan beberapa baris data RFM
rfm_df.sample(5)
```

```python
# Visualisasi
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

# Define color for the bar plots
color = "#72BCD4"

# Visualisasi berdasarkan Recency
sns.barplot(x="customer_id", y="recency", data=rfm_df.sort_values(by="recency").head(5), ax=ax[0], palette=[color])
ax[0].set_xlabel(None)
ax[0].set_xlabel('customer_id')
ax[0].set_ylabel("Recency (days)", fontsize=14)
ax[0].set_title("Top 5 Customers by Recency", fontsize=18)
ax[0].tick_params(axis='y', labelsize=12)
ax[0].set_xticks([])  # Remove x-tick labels to hide customer_id

# Visualisasi berdasarkan Frequency
sns.barplot(x="customer_id", y="frequency", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), ax=ax[1], palette=[color])
ax[1].set_xlabel(None)
ax[1].set_xlabel('customer_id')
ax[1].set_ylabel("Frequency", fontsize=14)
ax[1].set_title("Top 5 Customers by Frequency", fontsize=18)
ax[1].tick_params(axis='y', labelsize=12)
ax[1].set_xticks([])  # Remove x-tick labels to hide customer_id

# Visualisasi berdasarkan Monetary
sns.barplot(x="customer_id", y="monetary", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), ax=ax[2], palette=[color])
ax[2].set_xlabel(None)
ax[2].set_xlabel('customer_id')
ax[2].set_ylabel("Monetary", fontsize=14)
ax[2].set_title("Top 5 Customers by Monetary", fontsize=18)
ax[2].tick_params(axis='y', labelsize=12)
ax[2].set_xticks([])  # Remove x-tick labels to hide customer_id

# Set overall title
plt.suptitle("Top 5 Customers Based on RFM Parameters", fontsize=20)
plt.tight_layout()
plt.subplots_adjust(top=0.85)  # Adjust the top to make space for the suptitle
plt.show()
```

**Insight**
1. Recency (Kiri):
Pelanggan dengan nilai Recency tertinggi (bar di paling kanan) melakukan pembelian dalam waktu yang lebih lama dibandingkan pelanggan lainnya. Artinya, mereka lebih jarang berinteraksi baru-baru ini. Pelanggan di bar paling kiri, dengan nilai Recency yang sangat rendah, melakukan transaksi paling baru. Mereka adalah pelanggan yang baru-baru ini melakukan pembelian dan mungkin lebih aktif saat ini.

2. Frequency (Tengah):
Pelanggan di bar paling kiri memiliki frekuensi transaksi tertinggi. Ini menunjukkan bahwa pelanggan ini sering melakukan transaksi dalam rentang waktu yang ditentukan, menunjukkan loyalitas atau preferensi yang tinggi terhadap produk atau layanan yang ditawarkan. Semakin ke kanan, frekuensi transaksi menurun, artinya pelanggan ini bertransaksi lebih jarang.

3. Monetary (Kanan):
Pelanggan dengan nilai tertinggi (bar di paling kiri) menyumbang nilai transaksi terbesar, sehingga mereka dianggap sebagai pelanggan dengan nilai ekonomis tertinggi bagi bisnis. Pelanggan ini mungkin memberikan kontribusi terbesar dalam pendapatan. ```python

```python
# Menghitung jumlah transaksi per pelanggan
transaction_counts = all_df.groupby('customer_id')['order_id'].count().reset_index()
transaction_counts.columns = ['customer_id', 'transaction_count']

# Mengelompokkan pelanggan berdasarkan jumlah transaksi
def categorize_transactions(count):
    if count <= 5:
        return 'Low'
    elif count <= 20:
        return 'Medium'
    else:
        return 'High'

transaction_counts['transaction_category'] = transaction_counts['transaction_count'].apply(categorize_transactions)

# Gabungkan kembali dengan data pelanggan
all_df = pd.merge(all_df, transaction_counts[['customer_id', 'transaction_category']], on='customer_id', how='left')

# Tampilkan beberapa baris data hasil pengelompokan
all_df.head()
```

```python
# Binning berdasarkan nilai pembayaran
bins = [0, 100, 500, 1000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']

# Membuat kategori berdasarkan bin
all_df['payment_category'] = pd.cut(all_df['payment_value'], bins=bins, labels=labels, right=False)

# Tampilkan beberapa baris data hasil binning
all_df[['payment_value', 'payment_category']].sample(10)
```

```python
# Menghitung minggu dari tanggal pembelian
all_df['week_of_year'] = all_df['order_purchase_timestamp'].dt.isocalendar().week

# Tampilkan beberapa baris data hasil binning
all_df[['order_purchase_timestamp', 'week_of_year']].head()
```

```python
# Membuat grid untuk tiga grafik dalam satu baris
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Visualisasi 1: Distribusi Kategori Transaksi Pelanggan
sns.countplot(data=all_df, x='transaction_category', palette='Set2', ax=axes[0])
axes[0].set_title('Distribusi Kategori Transaksi Pelanggan')
axes[0].set_xlabel('Kategori Transaksi')
axes[0].set_ylabel('Jumlah Pelanggan')

# Visualisasi 2: Distribusi Kategori Pembayaran
sns.countplot(data=all_df, x='payment_category', palette='Set1', ax=axes[1])
axes[1].set_title('Distribusi Kategori Pembayaran')
axes[1].set_xlabel('Kategori Pembayaran')
axes[1].set_ylabel('Jumlah Transaksi')

# Visualisasi 3: Jumlah Transaksi per Minggu
weekly_transactions = all_df.groupby('week_of_year')['order_id'].count().reset_index()
sns.lineplot(data=weekly_transactions, x='week_of_year', y='order_id', marker='o', color='b', ax=axes[2])
axes[2].set_title('Jumlah Transaksi per Minggu')
axes[2].set_xlabel('Minggu ke-')
axes[2].set_ylabel('Jumlah Transaksi')
axes[2].grid(True)

# Menampilkan semua visualisasi
plt.tight_layout()
plt.show()
```

**Insight**
- Mayoritas pelanggan melakukan transaksi dengan nilai rendah, dan ada potensi untuk meningkatkan nilai transaksi dengan mendorong lebih banyak pelanggan untuk melakukan pembelian dalam kategori menengah hingga tinggi.
- Kategori pembayaran menunjukkan pola serupa, di mana sebagian besar pembayaran berada pada level rendah hingga sedang.
- Melihat fluktuasi mingguan, ada peluang untuk menganalisis faktor-faktor yang mendorong lonjakan transaksi pada minggu-minggu tertentu dan bagaimana cara memanfaatkan pola tersebut untuk meningkatkan penjualan.
